setting up the browser session and importing the needed packages/libraries/modules

```python
from selenium import webdriver as wd

driver=wd.Chrome("C:/Users/preet/Downloads/chromedriver")
driver.get("https://www.instagram.com/")
```

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait as wdw
from selenium.webdriver.support import expected_conditions as ec

from selenium.webdriver.common.keys import Keys

import time

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# defining wait object

wait=wdw(driver, 10)
```

#### 1 Analyzing the habits of the food bloggers

logging in using facebook

```python
# clicking on login using facebook button

login_with_facebook=driver.find_element_by_class_name("KPnG0")
login_with_facebook.click()

# waiting till the inputs become visible

wait.until(ec.visibility_of_element_located((By.ID, "email")))
wait.until(ec.visibility_of_element_located((By.ID, "pass")))

# filling the username and password

phone_input=driver.find_element_by_id("email")
password_input=driver.find_element_by_id("pass")

phone_input.send_keys("SAMPLE USERNAME")
password_input.send_keys("SAMPLE PASSWORD")

# clicking on login button

login_button=driver.find_element_by_id("loginbutton")
login_button.click()

# waiting till notification message is shown

wait.until(ec.visibility_of_element_located((By.CSS_SELECTOR, "button.aOOlW.HoLwm")))

# clicking on not now button

noty_button=driver.find_element_by_css_selector("button.aOOlW.HoLwm")
noty_button.click()
```

searching for food and fetching the names of the first 10 food handles

```python
# searching for food

search_input=driver.find_element_by_class_name("XTCLo")
search_input.send_keys("food")

# waiting till the results are visible

wait.until(ec.visibility_of_all_elements_located((By.CSS_SELECTOR, "._7UhW9.xLCgt.qyrsm.KV-D4.uL8Hv         ")))

# fetching the first 10 food handles

food_handles=driver.find_elements_by_css_selector("._7UhW9.xLCgt.qyrsm.KV-D4.uL8Hv         ")[:10]

# clearing the search input

search_input.clear()

# getting the food handle names from the web elements

for i in range(len(food_handles)):
    food_handles[i]=food_handles[i].get_attribute("innerHTML")
```

getting the followers of each of the 10 food handles

```python
followers=[]
```

```python
# iterating on the first 10 food handles, shown when we search for food

for i in range(10):
#     waiting till the search input is visible

    wait.until(ec.visibility_of_element_located((By.CLASS_NAME, "XTCLo")))

#     searching for food
    
    search_input=driver.find_element_by_class_name("XTCLo")
    search_input.send_keys("food")       
    
#     waiting till the (i+1)th food handle is visible
    
    wait.until(ec.visibility_of_element_located((By.XPATH, "//div[@role='none'][position()="+str(i+1)+"]")))  
    
#     clicking on the (i+1)th food handle
    
    food_handle=driver.find_element_by_xpath("//div[@role='none'][position()="+str(i+1)+"]")
    food_handle.click()
    
#     waiting till the followers link is loaded
    
    wait.until(ec.visibility_of_all_elements_located((By.XPATH, "//li[@class='Y8-fY ']/a/span")))
    
#     fetching the followers and appending to the followers array
    
    followers_count=driver.find_element_by_xpath("//li[@class='Y8-fY ']/a/span").get_attribute("innerHTML")   
    followers.append(followers_count)
    
#     going back to visit other food handles
    
    driver.back()
    
```

```python
# converting the elements of followers array to integers

for i in range(len(followers)):
    length=len(followers[i])
    
    lastChar=followers[i][length-1]
    if lastChar=='k' or lastChar=="m":
        if lastChar=="k":
            mFac=1000
        else:
            mFac=1000000
            
        followers[i]=int(float(followers[i][:(length-1)])*mFac)
    else:
        followers[i]=int(followers[i])
```

printing the top 5 food handles with their follower count(decreasing order of count of followers)

```python
np_followers=np.array(followers)
sort_order=np_followers.argsort()[::-1]
```

```python
np_top_5_followers=np_followers[sort_order][:5]
```

```python
top_5_handles=[]

for i in range(5):
    index=sort_order[i]
    
    top_5_handles.append(food_handles[index])
```

```python
for i in range(len(top_5_handles)):
    print(top_5_handles[i], np_top_5_followers[i])
```

getting the number of posts these handles have done in the past 3 days

```python
count_of_posts=[0 for i in range(5)]
```

```python
# check to make sure whether we have to search for food again or not

new_page_was_opened=True

# iterating on the first 10 food handles

for i in range(10):    
#     waiting till the search input is visible

    wait.until(ec.visibility_of_element_located((By.CLASS_NAME, "XTCLo")))
    
#     searching for food if a new page was opened
    
    if new_page_was_opened:
        search_input=driver.find_element_by_class_name("XTCLo")
        search_input.send_keys("food")       
        
#     waiting till the food handle to be clicked is visible
        
    wait.until(ec.visibility_of_element_located((By.XPATH, "//div[@role='none'][position()="+str(i+1)+"]")))    
    
#     fetching the food handle and its name
    
    food_handle=driver.find_element_by_xpath("//div[@role='none'][position()="+str(i+1)+"]")
    food_handle_name=driver.find_element_by_xpath("//div[@role='none'][position()="+str(i+1)+"]/a/div/div[position()=2]/div/div/div/div").get_attribute("innerHTML")        
    
#     exploring the food handle's page only if it is in the top 5 handles
    
    if not food_handle_name in top_5_handles:
        new_page_was_opened=False
        continue
    else:
        new_page_was_opened=True    
    
    food_handle.click()    
    
#     finding the count of posts the handle has done in the past 3 days and setting it at the respective index(wrt the top 5
#     handle ordering) of the count_of_posts array

    count=0
    index=top_5_handles.index(food_handle_name)
    
#     waiting for 3 seconds(to make sure the page has loaded properly and the content has changed)
    
    time.sleep(3)
    
    i=0
    
    while True:    
        pos1=str((i//3)+1)   
        
        pos2=str((i%3)+1)
        
#         waiting till the post to be selected is visible
        
        wait.until(ec.visibility_of_element_located((By.XPATH, "//div[@class='Nnq7C weEfm'][position()="+pos1+"]/div[position()="+pos2+"]")))
    
#         clicking on the post
    
        post=driver.find_element_by_xpath("//div[@class='Nnq7C weEfm'][position()="+pos1+"]/div[position()="+pos2+"]")
        post.click()
        
#         waiting till the time is visible
        
        wait.until(ec.visibility_of_element_located((By.CSS_SELECTOR, "time._1o9PC.Nzb55")))                        
        
#         getting the post time and current time

        time_element=driver.find_element_by_css_selector("time._1o9PC.Nzb55")        
        
        post_time=time_element.get_attribute("datetime")
        
        current_time=driver.execute_script("return new Date().getTime();")
        post_time=driver.execute_script("return new Date(arguments[0]).getTime();", post_time)
        
#         getting the minutes ago value when the post was posted
                
        minute_diff=driver.execute_script("return (arguments[0]-arguments[1])", current_time, post_time)
        minute_diff/=60000                             
        
#         closing the post
        
        close_button=driver.find_element_by_css_selector("div.Igw0E.IwRSH.eGOV_._4EzTm.BI4qX.qJPeX.fm1AK.TxciK.yiMZG button.wpO6b  ")
        close_button.click()
        
#         increasing the count if the post was posted 3 days ago or after it or we break 
        
        if minute_diff<=(3*24*60):
            count+=1    
        else:
            break   
            
        i+=1
        
#     setting the count_of_posts value for particular index
        
    count_of_posts[index]=count
    
#     going back to visit other food handles
    
    driver.back()
  
```

printing the number of posts, each of the top 5 handles have done in the past 3 days

```python
for i in range(len(count_of_posts)):
    print(top_5_handles[i], count_of_posts[i])
```

plotting a bar graph for better visualization

```python
plt.title("Posts done by top 5 food handles in past 3 days")

plt.xticks(rotation=30)
plt.xlabel("Food handle")
plt.ylabel("Number of posts in past 3 days")

plt.bar(top_5_handles, count_of_posts, width=0.5)
plt.show()
```

#### 2 Getting list of hashtags the friend should use in his posts

getting the hashtags used and their count for each of the top 5 food handles from their first 10 posts

```python
hashtags=[{} for i in range(5)]
```

```python
# visiting the page of each of the top 5 food handles(STEPS EXPLAINED ABOVE) and fetching the hashtags used and their
# frequency for each of the top 5 food handles

new_page_was_opened=True

for i in range(10):    
    wait.until(ec.visibility_of_element_located((By.CLASS_NAME, "XTCLo")))
    
    if new_page_was_opened:
        search_input=driver.find_element_by_class_name("XTCLo")
        search_input.send_keys("food")       
        
        
    wait.until(ec.visibility_of_element_located((By.XPATH, "//div[@role='none'][position()="+str(i+1)+"]")))    
    
    food_handle=driver.find_element_by_xpath("//div[@role='none'][position()="+str(i+1)+"]")
    food_handle_name=driver.find_element_by_xpath("//div[@role='none'][position()="+str(i+1)+"]/a/div/div[position()=2]/div/div/div/div").get_attribute("innerHTML")        
    
    if not food_handle_name in top_5_handles:
        new_page_was_opened=False
        continue
    else:
        new_page_was_opened=True
    
    food_handle.click()    

    count=0
    index=top_5_handles.index(food_handle_name)
    
    time.sleep(3)
    
    for i in range(10):    
        pos1=str((i//3)+1)   
        
        pos2=str((i%3)+1)
        
        wait.until(ec.visibility_of_element_located((By.XPATH, "//div[@class='Nnq7C weEfm'][position()="+pos1+"]/div[position()="+pos2+"]")))
    
        post=driver.find_element_by_xpath("//div[@class='Nnq7C weEfm'][position()="+pos1+"]/div[position()="+pos2+"]")
        post.click()
        
#         enclosing the statement in try-except as the post may not have any hashtags
        
        try:
            wait.until(ec.visibility_of_elements_located((By.CSS_SELECTOR, "a.xil3i")))               
        except:
            pass        
        
#         fetching all the hashtags of the post
        
        try:
            post_hashtags=driver.find_elements_by_css_selector("a.xil3i")
                        
#             updating the hashtags array
            
            for post_hashtag in post_hashtags:                
                hashtag_content=post_hashtag.get_attribute("innerHTML")[1:]
                
                hashtags[index][hashtag_content]=(hashtags[index].get(hashtag_content, 0)+1)
        except:
            pass
        
        wait.until(ec.visibility_of_element_located((By.CSS_SELECTOR, "div.Igw0E.IwRSH.eGOV_._4EzTm.BI4qX.qJPeX.fm1AK.TxciK.yiMZG button.wpO6b  ")))
        
        close_button=driver.find_element_by_css_selector("div.Igw0E.IwRSH.eGOV_._4EzTm.BI4qX.qJPeX.fm1AK.TxciK.yiMZG button.wpO6b  ")
        close_button.click()
    
    driver.back()
  
```

printing the hashtags used by each of the top 5 food handles and their count

```python
for i in range(len(hashtags)):
    print(top_5_handles[i], ":", end="\n\n")
    
    for j in hashtags[i]:
        print(j, hashtags[i][j])
        
    print()
```

printing the hashtags and their frequency collectively for the top 5 food handles

```python
hashtag_vs_count={}

for i in range(len(hashtags)):
    
    count+=len(hashtags[i])
    
    for j in hashtags[i]:
        hashtag_vs_count[j]=(hashtag_vs_count.get(j, 0)+hashtags[i][j])

for key, value in hashtag_vs_count.items():
    print(key, value)
```

creating a csv file with 2 cols(word and frequency)

```python
# getting the hashtag and its count in a 2d list

hashtag_vs_count_list=[]

for key, value in hashtag_vs_count.items():
    hashtag_vs_count_list.append([key, value])
```

```python
# creating a dataframe with 2 cols, word and frequency

df=pd.DataFrame(hashtag_vs_count_list, columns=["word", "frequency"])
df
```

```python
# creating a csv file from the dataframe

df.to_csv("WordVsFreq.csv", index=False)
```

printing the top 5 hashtags along with their frequencies, which were the most popular among these bloggers

```python
sort_order=(df["frequency"].values.argsort()[::-1])
```

```python
for i in range(5):
    index=sort_order[i]
    
    print(df["word"].iloc[index], df["frequency"].iloc[index])
```

plotting a pie chart of the top 5 hastags and their count percentage(among the top 5 only)

```python
plt.title("Top 5 hashtags used")
plt.axis("equal")

frequency_top_5=df["frequency"].values[sort_order][:5]
names_top_5=df["word"].values[sort_order][:5]

plt.pie(frequency_top_5, labels=names_top_5, autopct="%.2f")
plt.show()
```

#### 3 Calculating average Followers: Likes ratio for these top 5 food handles

getting the count of likes on the first 10 posts of these bloggers

```python
count_of_likes=[0 for i in range(5)]
```

```python
# visiting the page of each of the top 5 food handles(STEPS EXPLAINED ABOVE) and fetching the likes on their first 10 posts

new_page_was_opened=True

for i in range(10):    
    wait.until(ec.visibility_of_element_located((By.CLASS_NAME, "XTCLo")))
    
    if new_page_was_opened:
        search_input=driver.find_element_by_class_name("XTCLo")
        search_input.send_keys("food")       
        
        
    wait.until(ec.visibility_of_element_located((By.XPATH, "//div[@role='none'][position()="+str(i+1)+"]")))    
    
    food_handle=driver.find_element_by_xpath("//div[@role='none'][position()="+str(i+1)+"]")
    food_handle_name=driver.find_element_by_xpath("//div[@role='none'][position()="+str(i+1)+"]/a/div/div[position()=2]/div/div/div/div").get_attribute("innerHTML")        
    
    if not food_handle_name in top_5_handles:
        new_page_was_opened=False
        continue
    else:
        new_page_was_opened=True
    
    food_handle.click()    

    count=0
    index=top_5_handles.index(food_handle_name)    
    
    time.sleep(3)    
    
    for i in range(10):    
        pos1=str((i//3)+1)   
        
        pos2=str((i%3)+1)
        
        wait.until(ec.visibility_of_element_located((By.XPATH, "//div[@class='Nnq7C weEfm'][position()="+pos1+"]/div[position()="+pos2+"]")))
    
        post=driver.find_element_by_xpath("//div[@class='Nnq7C weEfm'][position()="+pos1+"]/div[position()="+pos2+"]")
        post.click()
        
#         fetching the likes/views(if the likes are not mentioned) on each post, after waiting till the element is visible
        
        try:
            wait.until(ec.visibility_of_element_located((By.XPATH, "//a[@class='zV_Nj']/span")))
            post_likes=driver.find_element_by_xpath("//a[@class='zV_Nj']/span").get_attribute("innerHTML")
        except:
#             waiting till the views are visible

            wait.until(ec.visibility_of_element_located((By.XPATH, "//span[@class='vcOH2']/span")))
            post_likes=driver.find_element_by_xpath("//span[@class='vcOH2']/span").get_attribute("innerHTML")
             
#         removing the commas and short-forms and converting likes/views to integer value
        
        post_likes=post_likes.replace(",", "")
        post_likes=post_likes.replace("k", "000")
        post_likes=post_likes.replace("m", "000000")
        
        post_likes=int(post_likes)
        
#         updating the count_of_likes array
        
        count_of_likes[index]+=post_likes
        
        close_button=driver.find_element_by_css_selector("div.Igw0E.IwRSH.eGOV_._4EzTm.BI4qX.qJPeX.fm1AK.TxciK.yiMZG button.wpO6b  ")
        close_button.click()
    
    driver.back()
 
```

getting the average count of likes 

```python
np_count_of_likes=np.array(count_of_likes)

np_average_likes=(np_count_of_likes//10)
```

getting the followers: likes ratio and printing it for the top 5 food handles

```python
np_average_followers_like_ratio=np.around((np_top_5_followers/np_average_likes), 2)

sort_order=np_average_followers_like_ratio.argsort()[::-1]

for i in range(len(np_average_followers_like_ratio)):
    index=sort_order[i]
    
    print(top_5_handles[index], np_average_followers_like_ratio[index])
```

plotting a bar graph for better visualization

```python
plt.title("Average followers to likes ratio for top 5 handles")

plt.xticks(rotation=30)
plt.xlabel("Food handle")
plt.ylabel("Average Followers:Likes ratio")

plt.bar(top_5_handles, np_average_followers_like_ratio, width=0.5)
plt.show()
```
